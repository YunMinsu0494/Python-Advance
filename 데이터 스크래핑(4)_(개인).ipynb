```python
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
driver = uc.Chrome()
driver.get("https://search.naver.com/search.naver?ssc=tab.news.all&where=news&sm=tab_jum&query=%EB%8C%80%EA%B5%AC%EC%9D%80%ED%96%89")
```

```python
# 네이버 대구은행 검색 뉴스 탭에서 1페이지 기사 제목과 요약 내용 가지고 오기
box = []
news = driver.find_element(By.CSS_SELECTOR, "ul.list_news._infinite_list").find_elements(By.CSS_SELECTOR, "li.bx")
for i in news:
    title = i.find_element(By.CSS_SELECTOR, "a.news_tit").get_attribute("title")
    content = i.find_element(By.CSS_SELECTOR, "div.dsc_wrap").text.strip()
    A = [title, content]
    if A not in box:
        box.append(A)

box
```

```python
# 방법 1(한번에 확 내려야 할때)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") #0부터 뉴스 본문 페이지 까지 스크롤 하기
```

```python
# 방법 2(조금씩 내려야 할때)
body = driver.find_element(By.TAG_NAME, "body")
for n in range(10):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)
```

```python
# 뉴스기사 100개 수집
box = []
while True:
    news = driver.find_element(By.CSS_SELECTOR, "ul.list_news._infinite_list").find_elements(By.CSS_SELECTOR, "li.bx")
    for i in news:
        title = (i.find_element(By.CSS_SELECTOR, "a.news_tit").get_attribute('title'))
        summary = (i.find_element(By.CSS_SELECTOR, "div.dsc_wrap").text.strip())
        A = [title, summary]
        if A not in box:
            box.append(A)
            if len(box) == 100:
                break
    if len(box) == 100:
        break
    print(f"{len(box)}개 뉴스기사 수집 완료")
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)
```

```python
import pandas as pd
df = pd.DataFrame(box, columns = ["제목", "내용"])
df.to_excel("네이버 뉴스기사.xlsx", index = False)
df
```

```python
# 사진 가지고 오기
import requests

idx = 0
for i in driver.find_elements(By.CSS_SELECTOR, "li._1OKV3UfxuI"):
    image = (i.find_element(By.TAG_NAME, "img"))
    src = (image.get_attribute("src"))
    if not src: 
        continue #src가 없으면 그냥 계속 넘어가라
    for j in ['.jpg', '.png', '.jpeg', '.gif']:
        if j in src:
            a = src.split("?")[0]
            f = open(f"{idx}{j}", 'wb')
            response = requests.get(src)
            f.write(response.content)
            f.close()
            idx += 1
```

#### 페이지가 모두 뜰때(HTML 한정)까지 최대 10초를 기다려라!

```python
# 셀레니움 개 사기 기술
box = []
for n in range(1, 720):
    driver.implicitly_wait(10) #페이지가 모두 뜰때까지 최대 10초를 기다려라!
    driver.get(f"https://finance.naver.com/item/sise_day.naver?code=005930&page={n}")
    html = BeautifulSoup(driver.page_source)
    table = html.find("table")
    table = pd.read_html(str(table))[0].dropna()
    box.append(table)
```

```python
samsung = pd.concat(box, ignore_index = True)
samsung = samsung.iloc[1:] # 2월 27일 기준으로 27일은 아직 형성중이기 때문에 제외
samsung.to_excel("삼성 주가 데이터.xlsx", index = False)
samsung
```
