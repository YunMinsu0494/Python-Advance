```python
import requests
from bs4 import BeautifulSoup
import time
import random
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver")
html = BeautifulSoup(url.text)
```

```python
def make_table(html):
    table = html.find("table", class_ = "type_2")
    table = pd.read_html(str(table))[0]
    del table["N"]
    del table["토론실"]
    table = table.dropna(subset = "종목명")
    return table
```

```python
items = html.find("table", class_ = "item_list").find_all("input", attrs = {"name" : "fieldIds"})

items_box = []
for i in items:
    items_box.append(i["value"])
items_box
```

```python
dic = {"menu" : "market_sum",
      "returnUrl" : "http://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1",
      "fieldIds" : items_box}

url = requests.post("https://finance.naver.com/sise/field_submit.naver",
             data = dic) # Network -> Headers에 들어가서 링크를 가져온 후, 앞에 만든 dic을 data로 불러온다.
# post: get 과 마찬가지로 불러오기 지만 추가적인 정보를 더할때 사용
html = BeautifulSoup(url.text)

table = make_table(html)
table.to_excel("시가총액.xlsx", index = True)
table
```

```python
# 코스피, 코스닥 모두 가져오기(열도 모든 열) -> 하나의 데이터 프레임 만들기.
url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1")
html = BeautifulSoup(url.text)

def make_table(html):
    table = html.find("table", class_ = "type_2")
    table = pd.read_html(str(table))[0]
    del table["N"]
    del table["토론실"]
    table = table.dropna(subset = "종목명")
    return table

kospi_page = int(html.find("td", class_ = "pgRR").find("a")["href"].split("=")[-1])
kospi_page
```

```python
# 코스피 전체(모든열)
total = []
for i in range(1,kospi_page + 1):
    dic = {"menu" : "market_sum",
      "returnUrl" : f"http://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={i}",
      "fieldIds" : items_box}
    url = requests.post("https://finance.naver.com/sise/field_submit.naver", data = dic)
    html = BeautifulSoup(url.text)
    tables = make_table(html)
    total.append(tables)
    print(f"코스피 {i}페이지 저장 완료")
    time.sleep(2)
```

```python
kospi_all = pd.concat(total, ignore_index = True)
kospi_all
```

```python
# 코스닥 전체(모든열)
url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page=1")
html = BeautifulSoup(url.text)

kosdaq_page = int(html.find("td", class_ = "pgRR").find("a")["href"].split("=")[-1])
kosdaq_page
```

```python
total2 = []
for i in range(1,kosdaq_page + 1):
    dic = {"menu" : "market_sum",
      "returnUrl" : f"http://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page={i}",
      "fieldIds" : items_box}
    url = requests.post("https://finance.naver.com/sise/field_submit.naver", data = dic)
    html = BeautifulSoup(url.text)
    tables = make_table(html)
    total2.append(tables)
    print(f"코스닥 {i}페이지 저장 완료")
    time.sleep(2)
```

```python
kosdaq_all = pd.concat(total2, ignore_index = True)
kosdaq_all
```

```python
# 코스피 코스닥 전체 합치기
df_all = pd.concat([kospi_all, kosdaq_all], ignore_index = True)
df_all
```

```python
# 데이터들 엑셀로 저장
kospi_all.to_excel("네이버 코스피 전체 + 모든 항목.xlsx", index = False)
kosdaq_all.to_excel("네이버 코스닥 전체 + 모든 항목.xlsx", index = False)
df_all.to_excel("네이버 코스닥,코스피 전체 + 모든 항목.xlsx", index = False)
```

```python
# 인터파크에 들어가서 팝업창을 닫고 이용후기로 들어가는 매크로
driver = uc.Chrome()
driver.get("https://tickets.interpark.com/goods/24015987")
time.sleep(3)
driver.find_element(By.CLASS_NAME, "popupCloseBtn.is-bottomBtn").click()
time.sleep(1)
A = driver.find_element(By.CSS_SELECTOR, "ul.navList").find_elements(By.CSS_SELECTOR, "li.navItem ")
for i in A:
    if "이용후기" in i.text:
        i.click()
```

```python
# 이용후기의 모든 제목과 내용, 별점을 불러온 뒤 엑셀로 저장
box = []
for i in driver.find_elements(By.CSS_SELECTOR, "li.bbsItem"):
    title = (i.find_element(By.CSS_SELECTOR, "strong.bbsTitleText").text.strip())
    content = (i.find_element(By.CSS_SELECTOR, "p.bbsText").text.strip())
    star = (i.find_element(By.CSS_SELECTOR, "div.prdStarIcon").get_attribute("data-star")) # get_attribute: 속성값을 가지고 오는코드
    name = (i.find_element(By.CSS_SELECTOR, "span.name").text.strip())
    date = (i.find_elements(By.CSS_SELECTOR, "li.bbsItemInfoList")[1].text)
    box.append([title, content, name, star, date])
```

```python
review = pd.DataFrame(box, columns = ["제목", "내용", "작성자", "별점", "날짜"])
review.to_excel("인터파크 리뷰.xlsx", index = False)
review
```

```python
# 1~10 페이지까지 클릭하는 코드
numbers = driver.find_element(By.CSS_SELECTOR, "ol.pageNumWrap").find_elements(By.CSS_SELECTOR, "li")

for i in numbers:
    i.click()
    time.sleep(2)
```

```python
# 처음~ 페이지까지 클릭하는 코드 // 전체 이용후기의 모든 제목과 내용, 별점을 불러온 뒤 엑셀로 저장
box = []

while True:
    numbers = driver.find_element(By.CSS_SELECTOR, "ol.pageNumWrap").find_elements(By.CSS_SELECTOR, "li")
    
    for n in numbers:
        n.click()
        time.sleep(2)

        for i in driver.find_elements(By.CSS_SELECTOR, "li.bbsItem"):
            title = (i.find_element(By.CSS_SELECTOR, "strong.bbsTitleText").text.strip())
            content = (i.find_element(By.CSS_SELECTOR, "p.bbsText").text.strip())
            star = (i.find_element(By.CSS_SELECTOR, "div.prdStarIcon").get_attribute("data-star"))
            name = (i.find_element(By.CSS_SELECTOR, "span.name").text.strip())
            date = (i.find_elements(By.CSS_SELECTOR, "li.bbsItemInfoList")[1].text)
            box.append([title, content, name, star, date])

    try:
        driver.find_element(By.CSS_SELECTOR, "a.pageNextBtn.pageArrow").click()
        time.sleep(2)
    except:
        break
```

```python
review_all = pd.DataFrame(box, columns = ["제목", "내용", "작성자", "별점", "날짜"])
review_all.to_excel("인터파크 리뷰_전체.xlsx", index = False)
review_all
```
