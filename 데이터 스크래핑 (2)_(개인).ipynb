```python
import requests
from bs4 import BeautifulSoup
import time
import random
import pandas as pd
```

```python
url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page=1")
html = BeautifulSoup(url.text)
```

```python
import warnings
warnings.filterwarnings("ignore")

def make_table(html):
    table = html.find("table", class_ = "type_2")
    table = pd.read_html(str(table))[0]
    del table["N"]
    del table["토론실"]
    table = table.dropna(subset = "종목명") #종목명에 누락값이 있는 열을 삭제하겠다
    return table
```

```python
kospi_page = int(html.find("td", class_ = "pgRR").find("a")["href"].split("=")[-1])
kospi_page
```

```python
# 퀴즈: 모든 코스피 페이지를 가져와서 하나의 데이터프레임 만들기!
total = []
for i in range(1,kospi_page + 1):
    url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={i}")
    html = BeautifulSoup(url.text)
    tables = make_table(html)
    total.append(tables)
    print(f"코스피 {i}페이지 저장 완료")
    time.sleep(2)
```

```python
kospi_all_page = pd.concat(total, ignore_index = True) # 구조가 똑같기 떄문에 열기준으로 합쳐야한다!
kospi_all_page["소속"] = "KOSPI"
kospi_all_page.to_excel("네이버 코스피 전체.xlsx", index = False)
kospi_all_page
```

```python
# 퀴즈 코스닥까지 모두 가져와서 코스피 + 코스닥을 합친 하나의 데이터프레임 만들기
# PER값이 0보다 크고, 10보다 작은 행만 추출해서 저평가 종목 발굴하기!
url = requests.get("https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page=1")
html = BeautifulSoup(url.text)
```

```python
kosdak_page = int(html.find("td", class_ = "pgRR").find("a")["href"].split("=")[-1])
kosdak_page
```

```python
def make_table2(html):
    table = html.find("table", class_ = "type_2")
    table = pd.read_html(str(table))[0]
    del table["N"]
    del table["토론실"]
    table = table.dropna(subset = "종목명")
    return table
```

```python
total2 = []
for i in range(1,kosdak_page + 1):
    url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.naver?sosok=1&page={i}")
    html = BeautifulSoup(url.text)
    tables = make_table2(html)
    total2.append(tables)
    print(f"코스닥 {i}페이지 저장 완료")
    time.sleep(2)
```

```python
kosdak_all_page = pd.concat(total2, ignore_index = True)
kosdak_all_page["소속"] = "KOSDAK"
kosdak_all_page.to_excel("네이버 코스닥 전체.xlsx", index = False)
kosdak_all_page
```

```python
df = pd.concat([kospi_all_page, kosdak_all_page], ignore_index = True)
df
```

```python
df2 = df[(df['PER'] > 0) & (df['PER'] < 10)]
df2
```

```python
# 전체 페이지를 불러와서 데이터 전처리도 처리!
# 3년전 < 2년전 < 1년전 < 배당금 발굴하기
url = requests.get("https://finance.naver.com/sise/dividend_list.naver?&page=1")
html = BeautifulSoup(url.text)
dividend = int(html.find("td", class_ = 'pgRR').find("a")['href'].split("=")[-1])
dividend
```

```python
def make_table3(html):
    table = html.find("table", class_ = "type_1 tb_ty")
    table = pd.read_html(str(table), header = 1)[0]
    table = table.dropna(subset = "종목명")
    return table
```

```python
total3 = []
for i in range(1, dividend + 1):
    url = requests.get(f"https://finance.naver.com/sise/dividend_list.naver?&page={i}")
    html = BeautifulSoup(url.text)
    table = make_table3(html)
    total3.append(table)
    print(f"네이버 배당{i}페이지 저장 완료")
    time.sleep(2)
```

```python
df = pd.concat(total3, ignore_index = True)

df["3년전"] = df["3년전"].replace("-", 0).astype("float")
df["2년전"] = df["2년전"].replace("-", 0).astype("float")
df["1년전"] = df["1년전"].replace("-", 0).astype("float")
df["배당금"] = df["배당금"].replace("-", 0).astype("float")
df.info()
```

```python
df
```

```python
A = df["3년전"] < df["2년전"]
B = df["2년전"] < df["1년전"]
C = df["1년전"] < df["배당금"]
df2 = df[A & B & C]
df2
```

```python
from datetime import datetime
today = str(datetime.today().date())

text = f"{today}기준 3년 연속 배당금이 상승한 기업 리스트\n\n"

for i in range(len(df2)):
    row = df2.iloc[i]
    name = row["종목명"]
    a,b,c,d = row["배당금"], row["1년전"], row["2년전"], row["3년전"]
    text += f'{name} : {int(d):,}원 → {int(c):,}원 → {int(b):,}원 → {int(a):,}원\n'

f = open("3년 연속 배당금 상승 기업.txt", 'wt')
f.write(text)
f.close()
```
