```python
import pandas as pd
df = pd.read_excel("Data/Naver Blog.xlsx", header = None)
df.columns = ["날짜", "저목", "링크"]
df
```

```python
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
driver = uc.Chrome()
driver.get("https://blog.naver.com/kurum_be/220260342448")
```

```python
# 블로그에서 현재 보고 있는창을 다른창으로 바꾼 다음 가지고 오는 방법
# 본문내용 텍스트 부분만 가지고 오기.
driver.switch_to.frame("mainFrame")
content = driver.find_element(By.ID, "postViewArea").text
content
```

```python
box = []
```

```python
# 나중에 집에가서 해봐야징
for i in range(len(box), len(df["링크"])): #처음에는 0 ~ 링크까지, 나중에 문제가 발견 되었을떄는 앞에 저장된 상태는 유지하고 계속 실행함
    driver.implicitly_wait(10)
    driver.get(df["링크"].iloc[i]) # 위에랑 연결

    try:
        driver.switch_to.frame("mainFrame")
    except:
        continue
        
    content = None
    try:
        content = driver.find_element(By.ID, "postViewArea").text.strip()
    except:
        try:
            content = driver.find_element(By.CSS_SELECTOR, "div.se-main-container").text.strip()
        except:
            pass
    box.append(content)
```
