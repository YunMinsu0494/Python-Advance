```python
# 데이터 스크래핑: 인터넷에서 정보를 가져오는 기술
import requests # 페이지 요청
url = requests.get("https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&q=%EB%A1%9C%EB%98%90")
```

```python
# 200: 정상( 가져오는데 충돌이 없었다.)
# 400번대: 존재하지 않는 페이지.
# 500번대: 비공개 페이지, 제한된 페이지, 오류
from bs4 import BeautifulSoup # html 소스코드를 파이썬으로 다룰 수 있게 변환

html = BeautifulSoup(url.text) # url.text를 파이썬으로 다룰 수 있는 객체로 바꿔줘!
html
```

```python
def make_numbers(html):
    box = html.find("div", class_ = "lottonum").find_all("span")
    del box[6:8] # box에서 필요없는 부분인 7, 8열 삭제
    
    numbers = [] # numbers 변수 지정
    for i in box: 
        numbers.append(int(i.text)) # box의 텍스트 부분만 하나씩 numbers 에 저장
    return numbers
```

```python
make_numbers(html)
```

```python
import time # 데이터 스크래핑에서는 데이터를 빨리 가져오는것이 좋지 않아서 time 을 통해서 일정 시간에 걸쳐서 가지고 올 수 있음

# 사용 예시
for i in range(1, 11):
    print(i)
    time.sleep(2)
```

```python
total = []
for i in range(1, 21):
    url = requests.get(f"https://search.daum.net/search?w=tot&rtmaxcoll=LOT&DA=LOT&q={i}%ED%9A%8C%20%EB%A1%9C%EB%98%90%EB%8B%B9%EC%B2%A8%EB%B2%88%ED%98%B8")
    html = BeautifulSoup(url.text) # 파이썬으로 변환
    numbers = make_numbers(html) # 앞의 maker_nubmers(html) 함수를 numbers로 이름 변경
    total.append(numbers) # total 이라는 변수에 저장
    print(f"{i}회 로또 데이터 저장 완료: {numbers}")
    time.sleep(2)
```

```python
# 연습문제 1
url1 = requests.get("https://search.naver.com/search.naver?ssc=tab.news.all&where=news&sm=tab_jum&query=%EB%8C%80%EA%B5%AC")
html1 = BeautifulSoup(url1.text)
html1
```

```python
news = html1.find("ul", class_ = "list_news _infinite_list").find_all("li", class_ = "bx")
len(news)
```

```python
total = []
for i in news:
    a = i.find("a", class_ = "news_tit")
    title = a["title"].strip()
    link = a["href"]
    total.append([title, link])
news = pd.DataFrame(total, columns = ["제목", "링크"])

news.to_excel("news.xlsx")
news
```

```python
total = [] 
for n in range(1, 21):
    url = requests.get(f"https://search.daum.net/search?w=tot&DA=JIM&rtmaxcoll=JIM&&q=%EC%97%B0%EA%B8%88%EB%B3%B5%EA%B6%8C%20720%2B%20{n}%ED%9A%8C%EC%B0%A8")
    html = BeautifulSoup(url.text)
    numbers = html.find("tr", class_ = "fst").text.split()[5:]
    total.append(numbers)
    print(f"{n}회 연금복권 데이터 저장완료 : {numbers}")
    time.sleep(2)
```
